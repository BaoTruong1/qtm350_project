# Plots 


### Setup

```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
employment = pd.read_csv("../data/cleaned_employment.csv")
growth = pd.read_csv("../data/cleaned_growth.csv")
gdp_adjusted = pd.read_csv("../data/cleaned_ppp.csv")
```

## Plot General Look

### Employment Basic Plots

```python
countries = employment['country'].unique()

for country in countries:
    country_data = employment[employment['country'] == country]
    plt.scatter(country_data['date'], country_data['employment'], label = f'{country} Data Points')
    plt.title(f'Scatter Plot: Employment vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('Employment')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### GDP Growth General Look

```python
countries = growth['country'].unique()

for country in countries:
    country_data = growth[growth['country'] == country]
    plt.scatter(country_data['date'], country_data['gdp_growth'], label = f'{country} Data Points')
    plt.title(f'Scatter Plot: GDP-Growth vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('GDP-Growth')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### GDP-Adjusted General Plots

```python
countries = gdp_adjusted['country'].unique()

for country in countries:
    country_data = gdp_adjusted[gdp_adjusted['country'] == country]
    plt.scatter(country_data['date'], country_data['ppp'], label = f'{country} Data Points')
    plt.title(f'Scatter Plot: GDP-Adjusted vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('GDP-Adjusted')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Graphs to Check for Linearity 

### Employment Linearity Plots

```python
countries_employment = employment['country'].unique()

for country in countries_employment:
    country_data = employment[employment['country'] == country]
    x = country_data['date'].values.reshape(-1, 1)
    y = country_data['employment'].values

    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)

    plt.scatter(x, y, label=f'{country} Data Points')
    plt.plot(x, y_pred, label=f'{country} Fitted Line')

    plt.title(f'Linearity Check: Employment vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('Employment')
    plt.legend()
    plt.grid(True)
    plt.show()
```

### GDP Growth Linearity Plots

```python
countries_growth = growth['country'].unique()

for country in countries_growth:
    country_data = growth[growth['country'] == country]
    x = country_data['date'].values.reshape(-1, 1)
    y = country_data['gdp_growth'].values

    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)

    plt.scatter(x, y, label=f'{country} Data Points')
    plt.plot(x, y_pred, label=f'{country} Fitted Line')

    plt.title(f'Linearity Check: Growth vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('Growth')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
countries_gdp_adjusted = gdp_adjusted['country'].unique()

for country in countries_gdp_adjusted:
    country_data = gdp_adjusted[gdp_adjusted['country'] == country]
    x = country_data['date'].values.reshape(-1, 1)
    y = country_data['ppp'].values

    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)

    plt.scatter(x, y, label=f'{country} Data Points')
    plt.plot(x, y_pred, label=f'{country} Fitted Line')

    plt.title(f'Linearity Check: Employment vs. Year ({country})')
    plt.xlabel('Year')
    plt.ylabel('Employment')
    plt.legend()
    plt.grid(True)
    plt.show()
```
